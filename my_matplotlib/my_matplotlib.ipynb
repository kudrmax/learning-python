```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# scatter plot

```python
x = np.random.random(20) * 100
y = np.random.random(20) * 100

plt.scatter(x, y, c='red', marker='.', s=200)
plt.show()
```

```python
x = np.random.random(1000) * 100
y = np.random.random(1000) * 100

plt.scatter(x, y, alpha=0.5)
plt.show()
```

# line plot

```python
x = [2000 + x for x in range(24)]
y = np.random.random(len(x)) * 100

plt.plot(x, y, c='red', lw=3, linestyle='--')
# plt.plot(x, y, 'r--', lw=3)
plt.show()
```

# bar plot

```python
x = ['C++', 'C', 'C#', 'Python', 'Go']
y = [100, 80, 10, 30, 65]

plt.bar(x, y, color='red', width=0.5, edgecolor='#000', lw=6)
# plt.bar(x, y, color='red', align='edge')
plt.show()
```

# histograms

```python
data = np.random.normal(20, 1.5, 1000)

plt.hist(data, color='r')
plt.show()
```

```python
data = np.random.normal(20, 1.5, 1000)

plt.hist(data, color='r', bins=5)
plt.show()
```

```python
data = np.random.normal(20, 1.5, 1000)

plt.hist(data, color='r', bins=[data.min(), 18, 22, data.max()])
plt.show()
```

```python
data = np.random.normal(20, 1.5, 1000)

plt.hist(data, color='r', cumulative=True)
plt.show()
```

# pie chart

```python
lang = ['C++', 'C', 'C#', 'Python', 'Go']
votes = [100, 80, 10, 30, 65]

plt.pie(votes, labels=lang, autopct='%.2f%%', startangle=90)
plt.show()
```

```python
lang = ['C++', 'C', 'C#', 'Python', 'Go']
votes = [100, 80, 10, 30, 65]
explodes = [0, 0, 0, 0, 0.3]

plt.pie(votes, labels=lang, explode=explodes)
plt.show()
```

# box plot

```python
data = np.random.normal(172, 8, 300)

plt.boxplot(data)
plt.show()
```

# customization

```python
years = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
income = [55, 56, 62, 61, 72, 72, 73, 75]

income_ticks = list(range(50, 81, 2))

plt.plot(years, income)
plt.title('Plot with title', fontsize=30, fontname='Arial')
plt.xlabel('Year')
plt.ylabel('Yearly income in USD')
plt.yticks(income_ticks, [f'${x}k' for x in income_ticks])
plt.show()
```

# figures and subplots

```python
data = np.random.normal(20, 1.5, 1000)

#   ,      -
fig, axs = plt.subplots(1, 2)
fig.suptitle('Title')
axs[0].hist(data)
axs[0].set_title('Title 1')
axs[1].hist(data, cumulative=True)
axs[1].set_title('Title 2')

plt.tight_layout()  #       

plt.savefig("output/test1.png", dpi=300)
plt.savefig("output/test2.png", dpi=300, transparent=True)
plt.savefig("output/test3.png", dpi=300, bbox_inches='tight')
plt.show()
```
